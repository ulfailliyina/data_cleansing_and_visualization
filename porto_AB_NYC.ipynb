```python
#df.host_id.unique().sum()
harga_per_wil=df[['neighbourhood_group','price']]
rata_wil=harga_per_wil.groupby('neighbourhood_group').agg('mean').reset_index()
```

```python
import seaborn as sns
```

```python
def barplot_value(x, y, title):
  chart = sns.barplot(x, y, ci = None)
  chart.set_title(title)
  for p in chart.patches:
        chart.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
            textcoords='offset points')
  return chart 
```

```python
barplot_value(rata_wil.neighbourhood_group, rata_wil.price, "Harga Rata-Rata Sewa per Malam Untuk \nSetiap Wilayah (neighbourhood_group)")
```

```python
count_wilayah=df.groupby(['neighbourhood_group'])['id'].count().nlargest(10)
#plt.figure(figsize=(14,6))
x=list(count_wilayah.index)
y=list(count_wilayah.values)
x.reverse()
y.reverse()
plt.title('Wilayah (neighbourhood_group) Paling Banyak Terdaftar',fontsize=14)
plt.xlabel('frekuensi',fontsize=12)
plt.ylabel('neighbourhood_group',fontsize=12)
plt.barh(x,y)
plt.show()
```

```python
count_bagianwilayah=df.groupby(['neighbourhood'])['id'].count().nlargest(10)
#plt.figure(figsize=(14,6))
x=list(count_bagianwilayah.index)
y=list(count_bagianwilayah.values)
x.reverse()
y.reverse()
plt.title('Wilayah Bagian (neighbourhood) Paling Banyak Terdaftar',fontsize=14)
plt.xlabel('frekuensi',fontsize=12)
plt.ylabel('neighbourhood',fontsize=12)
plt.barh(x,y)
plt.show()
```

```python
count_bagianwilayah=df.groupby(['neighbourhood'])['id'].count().nsmallest(20)
#plt.figure(figsize=(14,6))
x=list(count_bagianwilayah.index)
y=list(count_bagianwilayah.values)
x.reverse()
y.reverse()
plt.title('Wilayah Bagian (neighbourhood) Paling Sedikit Terdaftar',fontsize=14)
plt.xlabel('frekuensi',fontsize=12)
plt.ylabel('neighbourhood',fontsize=12)
plt.barh(x,y)
plt.show()
```
